```python
import pandas as pd
```

```python
df = pd.read_csv('scraping_big.csv')
```

```python
df.head()
```

```python
df.fillna('')
```

```python
df.subtitle = df.subtitle.astype(str)
df.title = df.title.astype(str)
```

```python
import json
df.to_json('temp.json', orient='records')
```

```python
df['tags'] = df['title'] + df['content']
```

```python
df.tags = df.tags.astype(str)
df['tags'] = df['tags'].apply(str.lower)
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x: x.strip(" ").lower().split())
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text:
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
df['url']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english', ngram_range=(1, 3))
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def recommend(uuid):
    idx = df[df['uuid'] == uuid].index[0]
    distances = similarity[idx]
    rec = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
    L = []
    for i in rec:
        L.append(df.iloc[i[0]])
    return L
```

```python
recommend('0026af85-18b8-42a1-966a-17ac9d7ce9ea')
```

```python
from joblib import dump
dump(similarity, 'content_based.joblib')
dump(df, 'article_df.joblib')
```
